```python
import os
os.environ['KERAS_BACKEND'] = 'theano'
```

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
```

```python
x_data = np.array([[1, 2], [2, 3], [3, 1], [4, 3], [5, 3], [6, 2]])
y_data = np.array([[0], [0], [0], [1], [1], [1]])
```

```python
x_data.shape, y_data.shape
```

```python
model = Sequential()
model.add(Dense(input_dim=2, output_dim=1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='sgd')
model.summary()
```

```python
model.fit(x_data, y_data, nb_epoch=200, verbose=2)
```

```python
print("2,1", model.predict_classes(np.array([[2, 1]])))
print("6,5", model.predict_classes(np.array([[6, 5]])))
```
